```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

products = []
prices = []
ratings = []
for i in range(1,2): 
    url = "https://www.flipkart.com/search?q=cat+food&sid=p3t%2C0mo%2Coiv&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=cat+food%7CCat+Food&requestId=b01f5190-d18c-4d62-b0c5-e77d99b953e4&as-searchtext=cat+&sort=popularity"+str(i)
    quote_page = requests.get(url) 
    print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
    for item in soup.findAll("div", attrs={"class":"_4ddWXP"}):
         page_no = str(i)            
         name=item.find("a",attrs={"class":"s1Q9rs"})
         price=item.find("div",attrs={"class":"_30jeq3"})
         rating=item.find("div",attrs={"class":"_3LWZlK"})     
         if name != None:
            if name != None:
                products.append(name.text)
            else:
                ratings.append(None) 
            if price != None:
                prices.append(price.text)
            else:
                ratings.append(None) 
            if rating != None: 
                ratings.append(rating.text)
            else:
                ratings.append(None) 
                
df=pd.DataFrame({"Product Name":products,"Prices":prices, "Ratings":ratings}) 
df.to_csv('cat.csv')
df 
```
