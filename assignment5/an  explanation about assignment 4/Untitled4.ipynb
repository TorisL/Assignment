```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

for i in range(1,4):
    url = "https://hk.trip.com/travel-guide/search/?keyword=%E7%8C%AB%E5%92%96&from=home"+str(i)
    html = urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    print(soup)
```

```python
titles = []
urls = []
rates = []
address = []
likes = []
for item in range(1,5): 
    url = "https://hk.trip.com/travel-guide/search/?keyword=%E7%8C%AB%E5%92%96&from=home"+str(item)
    quote_page = requests.get(url) 
    soup = BeautifulSoup(quote_page.content,'html.parser')
    
    for item in soup.findAll("div", attrs={"class":"gl-search-result_list"}):
         page_no = str(item)            
         title = item.find("a")
         url = item.a.get("href")
         rate = item.find("span",attrs={"class":"jsx-1438290371"})
         addr = item.find("div",attrs={"class":"gl-search-result-_list-position"})
         like = item.find("span",attrs={"class":"favourite-count"})
         if title != None and url != None and rate != None and addr != None and like != None:
                titles.append(title.text)
                urls.append(url)
                rates.append(rate.text) 
                address.append(addr.text)
                likes.append(like.text)
df=pd.DataFrame({"Shop Name":titles,"Shop Url":urls,"Ratings":rates,"Address":address,"Like":likes})           
df
```
